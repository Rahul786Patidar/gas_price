```python
# Calculate quartiles for each column
Q1 = data['Italy'].quantile(0.25)
Q3 = data['Italy'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify and remove outliers
outliers = ((data['Italy'] < lower_bound) | (data['Italy'] > upper_bound))
data_no_outliers = data[outliers]
```

```python
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.boxplot(data['Italy'])
plt.title('Before Removing Outliers')

plt.subplot(2, 1, 2)
plt.boxplot(data_no_outliers['Italy'])
plt.title('After Removing Outliers')

plt.tight_layout()
plt.show()
```

**Linear Regression**

```python
X=data[['Year']]
y=data[['Japan']]
```

```python
#import ML Algorithm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
# Make sure 'Year' is treated as a feature with a proper name
data['Year'] = data['Year'].astype(float)  # Convert 'Year' to float for compatibility
```

```python
data['Year']
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y ,test_size = 0.3 , random_state= 1)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
lr = LinearRegression()
```

```python
lr
```

```python
model = lr.fit(X_train , y_train)
model
```

```python
 target_year = lr.predict([[2009]])
```

```python
# Print the predicted gas price
print(f'Predicted Gas Price for 2009 is : {target_year}')
```

```python
print("intercept: ",model.intercept_)
```

```python
print("coef: ",model.coef_)
```

```python
print("RScore: ",model.score(X_test , y_test))
```

```python
predictions = lr.predict(X_test)
```

```python
predictions
```

```python
from sklearn import metrics
```

```python
print("MSE:",metrics.mean_squared_error(y_test,predictions))
```

```python
print("MAE:",metrics.mean_absolute_error(y_test,predictions))
```
